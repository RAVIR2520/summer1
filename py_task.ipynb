```python
import pandas as pd
import re
import numpy as np
```

```python
data = pd.read_csv('/home/shreya/Downloads/dataset1Offline.csv',skiprows=3,header=1)
```

```python
data.head()
```

```python


#drop some unnecessary columns
df=data.drop(['Unnamed: 5','Unnamed: 7','Unnamed: 10'],axis=1)
```

```python
df
```

```python
df['SO']=df['SO #'].str.extract('^([a-zA-z].*)',expand = False).str.strip()
```

```python
df['SO_key']=df['SO #'].str.extract('(^[0-9][0-9]-.*)',expand = False).str.strip()
```

```python
df['product_key']=df['SO #'].str.extract('(^[0-9][0-9][0-9]-.*)',expand = False).str.strip()
```

```python
df['Customer']=df['Customer'].fillna(method='ffill')
df['Customer PO']=df['Customer PO'].fillna(method='ffill')
df['SO']=df['SO'].fillna(method ='ffill')
df['SO_key']=df['SO_key'].fillna(method ='ffill')
```

```python
df
```

```python
df = df.dropna(subset = ['product_key'])
```
